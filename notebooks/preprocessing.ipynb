```python
import pandas as pd
df=pd.read_csv("raw_data/crop_data.csv")
```

```python
remove_list = [
    "Maida Atta","Wheat Atta","Moong Dal","Chana Dal","Masur Dal",
    "Tur Dal","Urd Dal","Moath Dal","Mustard Oil","Coconut Oil",
    "Gingelly Oil","Ghee","Sugar","Jaggery","Broken Rice",
    "Ground Nut Seed","Corriander seed","Coconut Seed",
    "Niger Seed","Castor Seed","Fish","Wood","Copra"
]
```
